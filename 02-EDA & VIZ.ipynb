```python
scaler=MinMaxScaler()
#fit scaler
scaler.fit(rc[['Cost Index', 'Year Composite']])
#transformed scaled data
rc_scaled=scaler.transform(rc[['Cost Index', 'Year Composite']])

#rever to dataframe
rc_scaled = pd.DataFrame(rc_scaled, columns=['Cost Index', 'Year Composite'])

rc_scaled.head()
```

```python
rc_scaled['Date']=rc.Date
```

```python
rc_scaled.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Line(x=rc_scaled['Date'], 
                      y=rc_scaled['Year Composite'],
                    mode='lines',
                    name='lines'))
fig.add_trace(go.Line(x=rc_scaled['Date'],
                      y=rc_scaled['Cost Index'],
                    mode='lines+markers',
                    name='lines+markers'))
fig.show()
```
